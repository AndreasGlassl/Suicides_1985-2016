```python
import pandas as pd
```

```python
suicides = pd.read_csv("../data/master_oecd.csv")
alcohol = pd.read_csv("../data/alcohol_per_capita2.csv")
poverty = pd.read_csv("../data/poverty.csv")
working_hours = pd.read_csv("../data/work_df.csv")
economic = pd.read_csv("../data/economic_info.csv")
```
