```python
import pandas as pd
```

```python
suicides = pd.read_csv("../data/master.csv")
suicides.head()
```

```python
suicides_a = (suicides
                .groupby(["continent","country","year","age"])
                .agg({"suicides_no":"sum",
                      "population":"sum",
                      "gdp_per_capita ($)":"mean"})
                .sort_values(["country","year"])
                .reset_index())

suicides_a = (suicides_a
                 .assign(s_rate=((suicides_a["suicides_no"]/suicides_a["population"])*100000)))
suicides_a["s_rate"]=round(suicides_a["s_rate"],2)
```

```python
suicides_a.head()
```

```python
#We create another subset, we "ignore" the partition by age group

suicides_f = (suicides_a
              .groupby(["continent","country","year"])
              .agg({"suicides_no":"sum",
                    "population":"sum",
                    "gdp_per_capita ($)":"mean"})
              .sort_values(["country","year"])
              .reset_index())

suicides_f = (suicides_f
                 .assign(s_rate=((suicides_f["suicides_no"]/suicides_f["population"])*100000)))
suicides_f["s_rate"]=round(suicides_f["s_rate"],2)
```

```python
suicides_f.head()
```

```python
suicides_f.to_csv("../data/master_grouped_ccy.csv")
suicides_a.to_csv("../data/master_grouped_plusage.csv")
```
