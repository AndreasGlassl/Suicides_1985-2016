```python
import pandas as pd
```

```python
suicides = pd.read_csv("../data/master_oecd.csv")
alcohol = pd.read_csv("../data/alcohol_per_capita2.csv")
poverty = pd.read_csv("../data/poverty.csv")
working_hours = pd.read_csv("../data/work_df.csv")
economic = pd.read_csv("../data/economic_info.csv")
```

```python
suicides.head()
```

```python
alcohol.head()
```

```python
df1 = suicides.merge(alcohol, how="left", on=["country","year"])
```

```python
poverty.head()
```

```python
df2 = df1.merge(poverty, how="left", on=["country","year"])
```

```python
df2.head()
```

```python
economic.head()
```

```python
eco = economic.drop(columns=["Unnamed: 0","age group","poverty rate","internet access"])
```

```python
eco["country"]=eco["country"].str.upper()
```

```python
df3 = df2.merge(eco, how="left", on=["country","year"])
```

```python
df3.head()
```

```python
working_hours.head()
```

```python
working_hours["country"]=working_hours["country"].str.upper()
```

```python
df4 = df3.merge(working_hours, how="left", on=["country","year"])
```

```python
df4[df4["year"]==2015]
```

```python
pension = pd.read_csv("../data/pension_df.csv")
inequality= pd.read_csv("../data/inequality_df.csv")
```

```python
pension.head()
```

```python
inequality
```

```python
df5 = df4.merge(pension, how="left", on=["country","year"])
```

```python
df6 = df5.merge(inequality, how="left", on=["country","year"])
```

```python
df6.to_csv("../data/final_df.csv", index=False)
```
