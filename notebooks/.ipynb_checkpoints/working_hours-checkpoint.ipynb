```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file = pd.read_csv("../data/working_hours_avg.csv", delimiter=";")
```

```python
file = file.drop(columns="Unnamed: 17")
```

```python
file =file.T
```

```python
file.columns = file.iloc[0]
```

```python
file = file.iloc[1:]
```

```python
file = file.reset_index(drop=False)
```

```python
for columns in file.columns:
    file[columns] = pd.to_numeric(file[columns])
```

```python
file["OECD countries"]
```

```python
fig, ax = plt.subplots(figsize=(15,10))
ax = sns.lineplot(data=file, x="index", y="Korea", ax=ax)
ax = sns.lineplot(data=file, x="index", y="Japan", ax=ax)
ax = sns.lineplot(data=file, x="index", y="OECD countries", ax=ax)
```
