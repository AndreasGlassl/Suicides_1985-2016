### Now we merge our poverty_gap_df with our suicides_rate_df to calculate the linear regression

```python
sui_r = pd.read_csv("../data/master_grouped_ccy.csv")
s_r = sui_r.drop(columns="Unnamed: 0")
```

```python
s_r.head()
```

```python
sui_po = s_r.merge(po_gap_tot, 
                      how="left", 
                      left_on=["country","year"], 
                      right_on=["LOCATION","TIME"])
```

```python
sui_po.head()
```

### We want to drop all the rows where LOCATION is NaN because those rows correspond to countries that are not oecd countries and we want focus only on oecd countries

```python
s_p = sui_po[sui_po["LOCATION"].notna()]
```

```python
s_p_cor = sm.add_constant(s_p)
s_p_cor = s_p_cor.rename(columns={"Value":"Poverty_gap"})
s_p_cor.head()
```

```python
time = [2013.00, 2014.00, 2015.00]
```

```python
s_p_11_15 = s_p_cor[s_p_cor["TIME"].isin(time)]
```

### Now we can finally draw the linear regression

```python
y = s_p_11_15["s_rate"]
X = s_p_11_15[["const","Poverty_gap"]]
lin_reg = sm.OLS(y,X)
results = lin_reg.fit()
results.summary()
```

```python
fig, ax = plt.subplots(figsize=(10,6))

sns.regplot(data=s_p_cor, x=s_p_11_15["Poverty_gap"], y=s_p_11_15["s_rate"], ax=ax)
```
